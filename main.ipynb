## Loading the Dataset

```python
print("Loading dataset...")

try:
    data = pd.read_csv("Amazon Sales data.csv")
except:
    print("Some error occured")

print("Dataset Sucessfully Loaded!")
```

## Data Cleaning & Filtering

### Checking for NULL values

```python
data.isnull().sum()
```

No null values

### Converting `Order Date` and `Ship Date` to <i>datetime</i>

```python
data["Order Date"] = pd.to_datetime(data["Order Date"])
data["Ship Date"] = pd.to_datetime(data["Ship Date"])
data.info()
```

### Extracting `Order Month` from `Order Date`

```python
data['Order Year'] = data['Order Date'].dt.year
data.head()
```

### Extracting `Order Year` from `Order Date`

```python
data['Order Month'] = data['Order Date'].dt.month_name()
data.head()
```

## Analysis

### Month-wise Sales Trend

#### Total Revenue generated per year

```python
month_wise_total_revenue = data.groupby('Order Month')['Total Revenue']
month_wise_total_revenue = month_wise_total_revenue.sum().sort_values().reset_index()
month_wise_total_revenue['Total Revenue'] /= 1000000

sns.barplot(x='Total Revenue', y='Order Month', data=month_wise_total_revenue, 
            hue='Order Month', palette='Blues')

plt.gca().invert_yaxis()
plt.title('Month-wise Sales Trend', fontsize=15)
plt.xlabel('Total Revenue (per $1M)', fontsize=12)
plt.ylabel('Month', fontsize=12)
plt.show()
```

<b>Key Oservations:-</b>
- Maximum `Total Revenue` is generated on month of <i>Feburary</i> (<i>~$25M</i>)
- Minimum `Total Revenue` is generated on month of <i>August</i> (<i>~$1M</i>)

#### Total Profit generated per year

```python
month_wise_total_profit = data.groupby('Order Month')['Total Profit']
month_wise_total_profit = month_wise_total_profit.sum().sort_values().reset_index()
month_wise_total_profit['Total Profit'] /= 1000000

sns.barplot(x='Total Profit', y='Order Month', data=month_wise_total_profit)

plt.gca().invert_yaxis()
plt.title('Month-wise Sales Trend', fontsize=15)
plt.xlabel('Total Profit (per $1M)', fontsize=12)
plt.ylabel('Month', fontsize=12)
plt.show()
```

<b>Key Oservations:-</b>
- Maximum `Total Profit` is generated on month of <i>Feburary</i> (<i>~$7M</i>)
- Minimum `Total Profit` is generated on month of <i>August</i> (<i><$1M</i>)

#### Units Sold per year

```python
month_wise_units_sold = data.groupby('Order Month')['Units Sold']
month_wise_units_sold = month_wise_units_sold.sum().sort_values().reset_index()
month_wise_units_sold['Units Sold'] /= 1000

sns.barplot(x='Units Sold', y='Order Month', data=month_wise_units_sold, 
            hue='Order Month', palette='Reds')

plt.gca().invert_yaxis()
plt.title('Month-wise Sales Trend', fontsize=15)
plt.xlabel('Units Sold (per 1K)', fontsize=12)
plt.ylabel('Month', fontsize=12)
plt.show()
```

<b>Key Oservations:-</b>
- Maximum `Units Sold` month of <i>July</i> (<i>~$75K</i>)
- Minimum `Units Sold` month of <i>March</i> (<i>~$15K</i>)

### Year-wise Sales Trend

```python
year_wise_total_revenue = data.groupby('Order Year')['Total Revenue'].sum().reset_index()
year_wise_total_revenue['Total Revenue'] /= 1000000

year_wise_total_profit = data.groupby('Order Year')['Total Profit'].sum().reset_index()
year_wise_total_profit["Total Profit"]/=1000000

year_wise_units_sold = data.groupby('Order Year')['Units Sold'].sum().reset_index()
year_wise_units_sold["Units Sold"]/=1000

# Generating Figure
plt.figure(figsize=(15,15)).suptitle("YEAR-WISE AMAZON SALES ANALYSIS", fontsize=18)

# Subplot-1 (Total Revenue per Year)
plt.subplot(2,2,1)
sns.barplot(x='Order Year', y='Total Revenue', data=year_wise_total_revenue, 
            hue='Order Year', palette='plasma')

plt.title('Total Revenue per Year', fontsize=15)
plt.ylabel('Revenue (per $1M)', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.grid(axis='y', color="k")

# Subplot-2 (Total Profit per Year)
plt.subplot(2,2,2)
sns.barplot(x='Order Year', y='Total Profit', data=year_wise_total_profit, 
            hue='Order Year', palette='magma')

plt.title('Total Profit per Year', fontsize=15)
plt.ylabel('Profit (per $1M)', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.grid(axis='y', color="k")

# Subplot-3 (Units Sold per Year)
plt.subplot(2,2,3)
sns.barplot(x='Order Year', y='Units Sold', data=year_wise_units_sold, 
            hue='Order Year', palette='cividis')

plt.title('Units Sold per Year', fontsize=15)
plt.ylabel('Units Sold (per 1K)', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.grid(axis='y', color="k")

# Subplot-4
plt.subplot(2,2,4)
plt.plot(year_wise_total_revenue['Total Revenue'], data=year_wise_total_revenue,
         marker="o", markersize=6)
plt.plot(year_wise_total_profit['Total Profit'], data=year_wise_total_profit,
         marker="o", markersize=6)
plt.legend(["Total Revenue", "Total Profit"])
plt.xticks(range(8),labels=year_wise_total_profit['Order Year'])
plt.title('Comparision b/w Total Profit & Total Revenue', fontsize=15)
plt.ylabel("USD (per $1M)", fontsize=12)
plt.xlabel("Year", fontsize=12)
plt.grid(True)
plt.show()
```

<b>Key Observations:-</b>

1. Plot-1 (Total Revenue/Year)
- Maximum revenue is generated in year <i>2012</i> (<i>>$30M</i>).
- Minimum revenue is generated in year <i>2011</i> (<i>~$11M</i>).
- Sudden rise in revenue in year <i>2012</i>,
- ...followed by gradual decrease in profit.

2. Plot-2 (Total Profit/Year)
- Maximum profit is generated in year <i>2012</i> (<i>>$9M</i>).
- Minimum profit is generated in year <i>2011</i> (<i><$3M</i>).
- Sudden rise in profit in year <i>2012</i>,
- ...followed by gradual decrease in profit.

3. Plot-3 (Units Sold/Year)
- Maximum units were sold in year <i>2012</i> (<i>~100K</i>).
- ...followed by year <i>2014</i> in second place (<i>~90K</i>).
- General range of `Units Sold` per year is between <i>40-70K</i>.

4. Plot-4 (Comparision b/w Total Profit & Total Revenue)
- **Strong relationship** between `Total Profit` & `Total Revenue`
- Rise in revenue and profit both in <i>2012</i>,
- ...followed by gradual decrease.
